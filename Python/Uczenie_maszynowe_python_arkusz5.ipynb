```python
#Zadanie 3 NumPy
import numpy

table = numpy.array([[3, 1, 2],
                    [9, 5, 7],
                    [4, 8, 6]])

row_sort = numpy.sort(table, axis=1)

column_sort = numpy.sort(table, axis=0)

flat_sort = numpy.sort(table.flatten())

# Wyświetlanie wyników
print(f"Tablica oryginalna:\n {table}")
print(f"Posortowana wierszami:\n {row_sort}")
print(f"Posortowana kolumnami:\n {column_sort}")
print(f"Tablica posortowana po spłaszczeniu:\n {flat_sort}")
```

```python
#Zadanie 6 NumPy
import numpy as np
import matplotlib.pyplot as plt

n = 100 
iterations = 50 
threshold = 30 

def random_walk():
    walk = np.zeros(n)
    
    for i in range(1, n):
        walk[i] = walk[i-1] + np.random.uniform(-1, 1)
    
    return walk

walk = random_walk()

plt.figure(figsize=(10, 6))
plt.plot(walk, label="Błądzenie losowe")
plt.title("Błądzenie losowe - 100 kroków")
plt.xlabel("Kroki")
plt.ylabel("Wartości")
plt.grid(True)
plt.legend()
plt.show()

min_value = np.min(walk)
max_value = np.max(walk)

print(f"Minimalna wartość: {min_value}")
print(f"Maksymalna wartość: {max_value}")

count_above_threshold = 0

for _ in range(iterations):
    walk = random_walk()
    if np.any(walk > threshold):  
        count_above_threshold += 1

print(f"Liczba przypadków, w których przekroczono {threshold}: {count_above_threshold} z {iterations}")

threshold_crossings = []

for _ in range(iterations):
    walk = random_walk()
    threshold_crossings.append(np.any(walk > threshold))

plt.figure(figsize=(10, 6))
plt.plot(range(1, iterations+1), threshold_crossings, 'o-', label="Przekroczono próg 30")
plt.title("Symulacje: Przekroczenie progu 30 w błądzeniu losowym")
plt.xlabel("Symulacja")
plt.ylabel("Przekroczono próg 30")
plt.grid(True)
plt.legend()
plt.show()
```
